```python
#In this code we would make and solve a tower of hanoi problem

#First we would create a function that would move the tower of hanoi
def Tower_of_Hanoi(n, source, auxiliary, destination):
  if n == 1:
    print(f"Moving disk 1 from {source} to {destination}")
    return
  Tower_of_Hanoi(n - 1, source, destination, auxiliary)
  print(f"Moving disk {n} from {source} to {destination}")
  Tower_of_Hanoi(n - 1, auxiliary, source, destination)

#defining parameters
source = "source"
auxiliary = "auxiliary"
destination = "destination"

#User input on how many disk the user wants and also making an error handler
try:
  Disk_number = int(input("Enter how many disk you want: ")) #user input on disk
  if Disk_number == 0: #Error handler on having no disk to move
    raise ValueError("There is no disk to move.")
  if Disk_number < 0:
    raise ValueError("Error! Input positive integers only!")
  Tower_of_Hanoi(Disk_number, source, auxiliary, destination)
except ValueError as e:
  print(str(e))

#Working as intended :)
```
