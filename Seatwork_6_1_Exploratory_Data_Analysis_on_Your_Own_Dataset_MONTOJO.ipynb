```python
filepath = '/content/serious-injury-outcome-indicators-2000-2022.csv'
```

```python
import pandas as pd
import numpy as np

injury_data = pd.read_csv(filepath)
injury_data
```

```python
import matplotlib.pyplot as plt

injury_data.plot()
plt.show()
```

```python
injury_data['Upper_CI'].plot()
plt.show()
```

```python
# Identifying the column names of the current data set
column_names_data = list(injury_data.columns)
print("The column names are:", column_names_data)
```

```python
# Identifying data types
injury_data_types = injury_data.dtypes
print("The data types of the serious injury data are:\n", injury_data_types)
```

```python
# Displaying the total number of records
Total_number_row = injury_data.shape[0]
Total_number_columns = injury_data.shape[1]

print("The total records by row:", Total_number_row)
print("The total records by columns:", Total_number_columns)
```

```python
# Displaying the first 20 records
injury_data.head(20)
```

```python
# Displaying the last 20 records
injury_data.tail(20)
```

```python
# Changing column name "Severity" to "Intensity"
injury_data.rename(columns = {'Severity' : 'Intensity'})
injury_data
```

```python
# Creating new column classification that display 'Catastrophic' if the value is 'Fatal', otherwise 'Non-fatal'
injury_data['Catastrophic'] = np.where(injury_data['Intensity'] == 'Fatal', 'Tragic', 'Non-tragic')
injury_data
```

```python
# Creating new dataframe 'In_danger' that gathers data with fatal
In_danger = injury_data[injury_data['Intensity'] == 'Fatal']
In_danger
```

```python
# Creating new dataframe 'Not_in_danger' that gathers data that excludes fatal in intensity
Not_in_danger = injury_data[injury_data['Intensity'] != 'Fatal']
Not_in_danger
```

```python
# Create a new dataframe "Children" that gathers data with age 0 to 14 years
Children = injury_data[(injury_data['Age'] == '0-14 years')]
Children
```

```python
# Create a new dataframe "Old" that gathers data with age of 75+ years
Old = injury_data[(injury_data['Age'] == '75+ years')]
Old
```

```python
# Create a new dataframe "All_ages" that gathers data of all ages
All_ages = injury_data[(injury_data['Age'] == 'All ages')]
All_ages
```

```python
# Using numpy to get the average Data_value and Lower_CI
print("The average of Data_value is:", np.mean(injury_data['Data_value']))

print("The average of Lower_Ci is:", np.sum(injury_data['Lower_CI'] / len(injury_data['Lower_CI'])))
```

```python
# Using numpy to get the median of Upper_CI
print("The median of Upper_CI is:", np.median(injury_data['Upper_CI']))
```

```python
# Using numpy to get middle value of Data_value

sorted_data_value = np.sort(injury_data['Data_value'])

middle_data_value = len(sorted_data_value) // 2

middle_Data_value = sorted_data_value[middle_data_value]

print("Middle value of Data_value:", middle_Data_value)
```

```python
# Using numpy to get the standard deviation of Upper_CI
print("The standard deviation of Upper_CI is:", np.std(injury_data['Upper_CI']))
```

# **Analysis**

In this dataset focusing on the severity of the injuries, the utilization of object data types shift towards qualitative information that involve textual description unlike the previous dataset that mainly deal with numerical data types like float and int. This dataset show the reason for the injury like assault and motor vehicle crashes and many more. It gives the upper and lower confidence interval that will determine how precise the estimates are.

Handling variables appropriately is needed because you can easily make mistakes. For an instance, in the 'Age' column, if you inputted "Children = injury_data[(injury_data['Age'] <= 14)]" here instead of "Children = injury_data[(injury_data['Age'] == '0-14 years')]", it would show an error saying TypeError: '<=' not supported between instances of 'str' and 'int' instead of running as intended.
