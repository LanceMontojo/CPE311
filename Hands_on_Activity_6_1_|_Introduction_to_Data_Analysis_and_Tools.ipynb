# **Mean**

```python
# Getting mean without importing the statistics module.
# First find the total number and thus use the length function.
length = len(salaries)
# After that add the numbers together.
adding_salaries = sum(salaries)
# Then, divide the total sum to the total number of the salary.
mean = adding_salaries / length
# Print the output
print("The mean of salaries is:", mean)
```

```python
# Confirming the mean with the statistics module.
import statistics

print("The mean of salaries is:", statistics.mean(salaries))

# Thus confirming the mean.
```

# **Median**

```python
# Getting median without importing the statistics module.
# First arrange data in ascending order.
sorting = len(salaries)
salaries.sort()

# If the number is odd, then the median is the middle point.
if sorting % 2 == 0:
  median1 = salaries[sorting//2]
  median2 = salaries[sorting//2 -1]
  median = (median1 + median2) / 2
# If the number is even, get the average of the two middle data points.
else:
  median = salaries[sorting//2]

print("The median of the salary is:", median)
```

```python
# Confirming the median using statistics import
import statistics

print("The median of the salary is:", statistics.median(salaries))
```

# **Mode**

```python
# In getting mode, we first need to sort the data and find the most common number that appear in the data.
# Getting mode without the statistics module but will import Counter
from collections import Counter
# Already have values for length
# Creating new variable named data that will count the occurence of each unique element in salaries and store it in data variable
data = Counter(salaries)

# Convert it into dictionary
getting_mode = dict(data)
# Finding which has the maximum value in the dictionary
mode = [multiple_salary for multiple_salary, value in getting_mode.items() if value == max(list(data.values()))]

#Checking if there multiple mode or no mode
if len(mode) == length:
  getting_mode = "No mode found!"
else:
  getting_mode = "The Mode is: " + ", ".join(map(str,mode))

print(getting_mode)
```

```python
# Confirming mode with statistics module
import statistics

print("The mode is:", statistics.mode(salaries))
```

# **Sample Variance**

```python
# In getting the sample variance, we first need to calculate and get the mean of the data set.
# Then subtract the mean from each data point in the salary set.
# Take the squared values obtained in the previous instruction
# Then add all the squared differences obtained.
# After that divide the values from the total number of the salary set.
# Getting the sample variance without statistics module
# Already have values for mean and length
Sample_Variance = sum((i - mean) ** 2 for i in salaries) / (length - 1)

print("The sample variance of the salary is:", Sample_Variance)
```

```python
# Checking using statistics module
import statistics

print("The sample variance of the salary is: % s"%(statistics.variance(salaries)))
```

# **Sample Standard Deviation**

```python
# Getting the sample standard deviation without the statistics module
# Since we got the sample variance already we only need to square root it to get the sample standard deviation.
Sample_Standard_Deviation = Sample_Variance ** 0.5
print("The sample standard deviation is:", Sample_Standard_Deviation)
```

```python
# Confirming the Sample Standard Deviation with the statistics module
import statistics

print("The sample standard deviation is:", statistics.stdev(salaries))
```

# **Coefficient of variation**

```python
# To get coefficient of variation we need to divide standard deviation to mean and multiply the result to 100
import statistics

coefficient_of_variation = (statistics.stdev(salaries) / statistics.mean(salaries)) * 100
print("The coefficient of variation is:", coefficient_of_variation)
```

# **Interquartile Range**

```python
import statistics
# In getting interquartile range, we first need to order data, followed by finding the median
sorted_salary = sorted(salaries)

# Then calculate the median of both the lower and upper half of the data
quartile1 = statistics.quantiles(sorted_salary, n = 4)[0]
quartile3 = statistics.quantiles(sorted_salary, n = 4)[-1]

# Lasty the IQR is the difference between the upper and lower medians.
Interquartile_Range = quartile3 - quartile1
print("The first quartile:", quartile1, "\nThe third quartile:", quartile3, "\nThe Interquartile Range:", Interquartile_Range)
```

# **Quartile coefficient of dispersion**

```python
# To get the Quartile coefficient of dispersion we need to subtract the third quartile to the first quartile
# And then divide it to the sum of the third quartile and the first quartile
# Let's copy the interquartile range code a bit
sorted_salary = sorted(salaries)
quartile1 = statistics.quantiles(sorted_salary, n = 4)[0]
quartile3 = statistics.quantiles(sorted_salary, n = 4)[-1]

# Getting Quartile coefficient of dispersion
QCoD = (quartile3 - quartile1) / (quartile3 + quartile1)
print("The quartile coefficient of dispersion:", QCoD)
```
